```python
# Solve a generalized assingment problem between 300 Ts and 300 Cs
# setting random weights on the assignments such that we maximize the
# total assignment weigthed objective
from gurobipy import *
from numpy import *
```

```python
# Set the data

numT = 300
numC = 300

Assignment = random.random((numT,numC))
```

```python
# Initialize the model
m=Model("Assignment")
```

```python
# Define the binary variables Xs
X = []
for t in range(numT):
    X.append([])
    for c in range(numC):
        X[t].append(m.addVar(vtype=GRB.BINARY,name="X%d%d"% (t, c)))
```

```python
# Set the optimization direction
m.update()
m.modelSense = GRB.MAXIMIZE
```

```python
# Define the assignment constraints
constraintT = []
constraintC = []

# One variable assigned per column
for t in range(numT):
    constraintT.append(m.addConstr(quicksum(X[t][c] for c in range(numC))
                                   == 1 ,'constraintT%d' % t))

# One variable assigned per row
for c in range(numC):
    constraintT.append(m.addConstr(quicksum(X[t][c] for t in range(numT))
                                   == 1 ,'constraintC%d' % t))
```

```python
# Define objective
m.setObjective(quicksum(quicksum([X[t][c]*Assignment[t][c] 
                        for c in range(numC)]) 
                        for t in range(numT)))

m.update()
```

```python
# Solve the optimization problem
m.optimize()
```

```python
# Print runtime and solution

print 'runtime is',m.Runtime, 's'

# Print solution
if m.status == GRB.Status.OPTIMAL:
    for t in range(numT):
        for c in range(numC):
            if X[c][t].x > 0.5:
                print('%g <-> %g' % (c, t))
```
